```python
#first i want to create our audio to speech model so we can use it in our bigger model

import sounddevice as sd #sounddevice is used for recording audio from the microphone
import whisper #speech to text model
import scipy.io.wavfile as wav #this is a library for saving audio files
from whisper.audio import load_audio
import tempfile

#this is a pretrained speech to text model from OpenAI (we're using the base model)
model = whisper.load_model("base")

#lets create a function to record audio from the user
def record_audio(duration, sample_rate=16000): #function to record audio with a duration and sample rate
    print("Recording...") #this is the message that will be printed when the recording starts
    audio = sd.rec(int(duration * sample_rate), samplerate=sample_rate, channels=1, dtype='float32')
    sd.wait()  #waits till the step before this is done
    print("Recording finished.") #printing the message when the recording is finished
    return audio, sample_rate #returns the audio and sample rate

#saving the file
def save_audio(audio, sample_rate, file_path): 
    wav.write(file_path, sample_rate, (audio * 32767).astype(np.int16))
    print(f"Audio saved to {file_path}")

def transcribe_audio(audio, sample_rate):
    #saving the audio to a temporary file
    #temp_file = os.path.abspath("temp_audio.wav")
    #print(f"Saving audio to {temp_file}")
    #save_audio(audio, sample_rate, temp_file)

   # temp_file = os.path.join(tempfile.gettempdir(), "temp_audio.wav")
    
    temp_file = r"C:\Users\Moataz Elsherity\Desktop\capstone 3\temp_audio.wav"
    print(f"Saving audio to: {temp_file}")
    save_audio(audio, sample_rate, temp_file)

    try:
        audio_data = load_audio(temp_file)
        print("Whisper successfully read the file!")
    except Exception as e:
        print(f"Whisper cannot read the file: {e}")

    #transcribing the audio 
    result = model.transcribe(temp_file) 
    return result["text"]

#lets test if it works!
duration = 10  # Duration in seconds
audio, sample_rate = record_audio(duration)
transcription = transcribe_audio(audio, sample_rate)
print("Transcription:", transcription)
```
